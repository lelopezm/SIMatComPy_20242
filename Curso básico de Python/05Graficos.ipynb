# Módulo **Matplotlib**

Esta es una librería gráfica para Python. Primero se importa la librería. Así:

```python
import matplotlib.pyplot as plt
import numpy as np
```

Generalmente para graficar una función creamos un *arreglo* para la variable independiente $x$, por ejemplo, con *linspace*, luego definimos la variable dependiente $y=f(x)$ y finalmente le decimos que grafique $(x,y)$ con la instrucción **plot**. 

Vamos a graficar la parábola $y=x^2$ con $x\in[-2,2]$, tabulando 100 puntos.

```python
a=-2
b=2
n=100
x=np.linspace(a,b,n)
y=x**2
plt.plot(x,y)
plt.show()
```

A continuación dejo algunos ejemplos de gráficas de funciones reales donde se realiza una configuración en los ejes, colores, etc.

### Ejemplo 1. 
Iniciamos graficando las funciones $f(x)=\cos{x}$ y $g(x)=\ln{(x+1)}$, con $x\in[0,7]$

```python
#Definimos la variable independiente
ax=0
bx=7
n=100
x = np.linspace(ax, bx, n)
```

```python
#Definimos las funciones f y g
def f(x):
    return np.cos(x)

def g(x):
    return np.log(x+1)

#ahora definimos las variables dependientes
y1 = f(x)
y2 = g(x)
```

```python
#Realizamos los gráficos
#Usando colores RGB (en porcentaje) y tamaño de linea 2 pt.
plt.plot(x,y1,'--',color=(0.8,0,0.5),linewidth=2,label='$y=\cos{x}$')
plt.plot(x,y2,color=(0,0.6,0),linewidth=2,label='$y=\ln{(x+1)}$')
plt.title('Funciones trigonométrica y logarítmica')
plt.xlabel('x') #nombre del eje horizontal
plt.ylabel('y') #nombre eje vertical
plt.legend(loc="lower right") #leyeda y localización: upper: aariba, lower: abajo
plt.grid(True) #mallado
plt.show()
```

### Ejemplo 2. 

Gráfica de una parábola $f(x)=ax^2+bx+c$ con sus elementos.

```python
def f(x):
  y=a*x**2+b*x+c
  return y

# Valores de los coeficientes y=ax^2+bx+c
a=2
b=-np.sqrt(15)
c=-4
vx=-b/(2*a)
k=3
#dominio [x0,xf]
x0=vx-k
xf=vx+k
X=np.linspace(x0,xf,500)
Y=f(X)
ymax=np.max(Y)
ymin=np.min(Y)
plt.plot(X,Y,'#0B0FF2',linewidth=1.5,label='$f(x)=ax^2+bx+c$')
plt.scatter(vx,f(vx),color='#D90EF2',linewidths=2.0, zorder=5, label='Vértice')
plt.text(vx,f(vx),'$V$', color='#D90EF2',fontsize=12, ha='right',va='bottom', fontweight='bold')
plt.axhline(0, color='black', linewidth=1) #eje x
plt.text(xf,0,'$x$', fontsize=10, ha='left', va='top') #etiqueta del eje x
plt.axvline(0, color='black', linewidth=1) #eje y
plt.text(0,ymax+k,'$y$', fontsize=10, ha='right', va='bottom') #etiqueta del eje x
plt.grid() #mallado
#plt.xticks(np.arange(x0,xf,2)) #control mallado horizontal
#plt.yticks(np.arange(ymin,ymax,1.5)) #control mallado vertical
plt.legend()
plt.xlim([x0,xf])
plt.ylim([ymin-k,ymax+k])
plt.show()
```

### Ejemplo 3.

Gráfica de $f(x)=\sin{x}$ y $g(x)=\cos{x}$ con $x\in[0,2\pi]$.

```python
def f(x):
  return np.sin(x)

def g(x):
  return np.cos(x)

#dominio [x0,xf]
x0=0
xf=2*np.pi
X=np.linspace(x0,xf,500)
Y1=f(X) #función seno
Y2=g(X) #función coseno
plt.plot(X,Y1,'#0B0FF2',linewidth=1.5,label='$f(x)=sen(x)$')
plt.plot(X,Y2,'#F2890B',linewidth=1.5,label='$f(x)=cos(x)$')
plt.scatter(np.pi/4,f(np.pi/4),color='#1CF20B',linewidths=2.0, zorder=5)
plt.scatter(5*np.pi/4,f(5*np.pi/4),color='#F20B35',linewidths=2.0, zorder=5)
plt.text(0,0,'$0$', color='black',fontsize=10, ha='left',va='top', fontweight='bold')
plt.text(np.pi/2,0,'$\dfrac{\pi}{2}$', color='black',fontsize=10, ha='right',va='top', fontweight='bold')
plt.text(np.pi,0,'$\pi$', color='black',fontsize=10, ha='right',va='top', fontweight='bold')
plt.text(3*np.pi/2,0,'$\dfrac{3}{2}\pi$', color='black',fontsize=10, ha='left',va='top', fontweight='bold')
plt.text(2*np.pi,0,'$2\pi$', color='black',fontsize=10, ha='left',va='top', fontweight='bold')
#puntos de intersección
plt.text(np.pi/4,0,'$\dfrac{\pi}{4}$', color='red',fontsize=10, ha='right',va='top', fontweight='bold')
plt.text(0,np.sqrt(2)/2,'$\dfrac{\sqrt{2}}{2}$', color='red',fontsize=10, ha='right',va='top', fontweight='bold')
plt.text(5*np.pi/4,0,'$\dfrac{5}{4}\pi$', color='red',fontsize=10, ha='right',va='top', fontweight='bold')
plt.text(0,-np.sqrt(2)/2,'$\dfrac{-\sqrt{2}}{2}$', color='red',fontsize=10, ha='right',va='top', fontweight='bold')
plt.plot([np.pi/4,np.pi/4],[0,np.sqrt(2)/2],'--r',linewidth=0.7)
plt.plot([0,np.pi/4],[np.sqrt(2)/2,np.sqrt(2)/2],'--r',linewidth=0.7)
plt.plot([5*np.pi/4,5*np.pi/4],[0,-np.sqrt(2)/2],'--r',linewidth=0.7)
plt.plot([0,5*np.pi/4],[-np.sqrt(2)/2,-np.sqrt(2)/2],'--r',linewidth=0.7)
#ejes coordenados
plt.axhline(0, color='black', linewidth=1.5) #eje x
plt.text(xf+0.3,0,'$x$', fontsize=10, ha='left', va='top') #etiqueta del eje x
plt.axvline(0, color='black', linewidth=1.5) #eje y
plt.text(0,1,'$y$', fontsize=10, ha='right', va='bottom') #etiqueta del eje x
plt.grid() #mallado
plt.xticks(np.arange(x0,xf,np.pi/2)) #control mallado horizontal
plt.legend(loc="lower right")
plt.xlim([-0.8,7]) #configuración horizontal
plt.show()
```

Finalmente, para pintar una **dispersión de puntos** usamos el comando *scatter*.

```python
Xt = np.linspace(0,4,10)
Yt = 4*Xt - Xt**2
plt.scatter(Xt,Yt,marker='D',s=10,color='#1CF20B') #marker se usa para la forma (marca), s: para el tamaño
plt.xlabel('x')
plt.ylabel('y')
plt.title('Dispersión de datos')
plt.axis('equal')
plt.grid()
plt.show()
```
